```python
import pandas as pd

df=pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Creating the inputs and target variable. 

# Inputs- Independent variable.

# Target- Dependent variable.
```

```python
inputs = df.drop(['Id','Species'],axis='columns')
target = df['Species']
```

```python
inputs.head()
```

# Label encoding for converting the values of target column.

```python
#Importing sklearn.

import sklearn
```

```python
#Importing labelencoder from sklearn library.

from sklearn.preprocessing import LabelEncoder
```

```python
#Object creation of labelEncoder.

le_target = LabelEncoder()
```

```python
target = le_target.fit_transform(target)
```

```python
target
```

```python
target= pd.DataFrame(target)
```

```python
target.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
clf2= model.fit(inputs,target)
```

```python
model.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(80,90))

tree.plot_tree(clf2,filled=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(inputs, target, test_size=0.25)
```

```python
len(X_train)
```

```python
len(Y_test)
```

```python
clf= model.fit(X_train,Y_train)
```

```python
predicted = model.predict(X_test)
```

```python
expected = Y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cr= confusion_matrix(predicted, expected)
```

```python
cr
```

```python
score = model.score(X_test, Y_test)
print(score)
```

```python
plt.figure(figsize=(80,90))
tree.plot_tree(clf,filled=True)
```
